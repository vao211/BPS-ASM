```python
import pandas as pd
import matplotlib.pyplot as plt

#load predictions
lgb_results = pd.read_csv('prediction/lgb_predictions.csv')
xgb_results = pd.read_csv('prediction/xgb_predictions.csv')

#load value
y_val = lgb_results['y_val']#real value
y_pred_lgb = lgb_results['y_pred_lgb']
y_pred_xgb = xgb_results['y_pred_xgb']

#scatter plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Scatter plot LightGBM
ax1.scatter(y_val, y_pred_lgb, color='blue', alpha=0.1, label='LightGBM', s=10)
ax1.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=2, label='Ideal')
ax1.set_title('LightGBM (RMSE: 469.62, RMSPE: 0.0838)', fontsize=12)
ax1.set_xlabel('Actual Sales', fontsize=10)
ax1.set_ylabel('Predicted Sales', fontsize=10)
ax1.legend()
ax1.grid(True)

# Scatter plot XGBoost
ax2.scatter(y_val, y_pred_xgb, color='red', alpha=0.1, label='XGBoost', s=10)
ax2.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=2, label='Ideal')
ax2.set_title('XGBoost (RMSE: 484.72, RMSPE: 0.078)', fontsize=12)
ax2.set_xlabel('Actual Sales', fontsize=10)
ax2.set_ylabel('Predicted Sales', fontsize=10)
ax2.legend()
ax2.grid(True)


plt.tight_layout()
fig.suptitle('Comparison of LightGBM and XGBoost Predictions', fontsize=14, y=1.05)
plt.savefig('model_comparison_separate_plots.png', dpi=480)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

lgb_results = pd.read_csv('prediction/lgb_predictions.csv')
xgb_results = pd.read_csv('prediction/xgb_predictions.csv')
y_val = lgb_results['y_val']
y_pred_lgb = lgb_results['y_pred_lgb']
y_pred_xgb = xgb_results['y_pred_xgb']

plt.figure(figsize=(10, 6))
# ScatterLightGBM
plt.scatter(y_val, y_pred_lgb, color='blue', alpha=0.3, label='LightGBM', s=50)
# Scatter XGBoost
plt.scatter(y_val, y_pred_xgb, color='red', alpha=0.3, label='XGBoost', s=50)
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=2, label='Ideal')

plt.title('Comparison of LightGBM and XGBoost Predictions', fontsize=14)
plt.xlabel('Actual Sales', fontsize=12)
plt.ylabel('Predicted Sales', fontsize=12)
plt.legend()
plt.grid(True)

plt.savefig('model_comparison_plot.png')
plt.show()
```
